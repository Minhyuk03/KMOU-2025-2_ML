LAB 8-1 Fasion-Mnist데이터 분류

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#데이터
fasion_mnist = tf.keras.datasets.fashion_mnist

(x_train,y_train),(x_test,y_tesy) = fasion_mnist.load_data()

#이미지 확인
num = x_train[1001]
for i in range(28):
  for j in range(28):
    print("{:4d}".format(num[i][j]),end = '')
  print()

print("x_train 데이터 형태",x_train.shape)
print("x_train[0] 데이터 형태",x_train[0].shape)
print("y_train 데이터 형태",y_train.shape)

#입력값 정규화
x_train = x_train / 255
x_test = x_test / 255

```

```python
data = np.arange(1,11).reshape(2,5).astype(np.float32)
print(data)

layer = tf.keras.layers.Dropout(.3, input_shape = (2, ))
outputs = layer(data, training = True)
print(outputs)

#드롭아웃계층을 가진 신경망 모델
model = keras.models.Sequential( [
      keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(128, activation='relu'),  # 첫 번째 은닉층: 128개 노드
    keras.layers.Dense(32, activation='relu'),   # 두 번째 은닉층: 32개 노드
    keras.layers.Dense(10, activation='softmax') # 출력층: 10개 클래스
])

#compile 모델 설정
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
#fit 모델학습
history = model.fit(x_train, y_train, epochs = 10, validation_data = (x_train, y_train))

pd.DataFrame(history.history).plot(figsize = (10,6))

model.summary()
```
